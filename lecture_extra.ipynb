```python
import socket 
import threading

clients = {}

def recieve(s:socket.socket, addr):
    while True:
        a = s.recv(10000)
        if len(a) == 0: #socket quiet
            break
        for i, s in clients.items():
            if i != addr:
                s.send(a)
    del clients[addr]


s = socket.socket()
s.bind(('0.0.0.0', 10000))
s.listen()

while True:
    try:
        a, b = s.accept()
        print(b, 'connected')
        clients[b] = a
        th = threading.Thread(target = recieve, args=(a,b))
        th.start()
    except KeyboardInterrupt:
        break
    except Exception as e:
        break

for i, sock in clients.items():
    sock.close()
s.close()
```
