```python
class Print_Data(Exception):
    pass

def accumulate():
    s = 0
    while True:
        try:
            s += yield s
        except Print_Data:
            print(s)

a = iter(accumulate())
next(a)

while True:
    match input().split():
        case ['put', x]:
            b = a.send(float(x))
            print(f'current sum is {b}')
        case ['print']:
            a.throw(Print_Data())
        case ['exit']:
            break
```

```python
def f(x):
    return x**2

def decor(n):
    def dec1(fun)
        def new_fun(x):
            res = fun(x)
            if res > n:
                return float('inf')
            return res
        return new_fun
    return dec1

print(f(1), f(11))
f = decor(f)
print(f(1), f(11))

@decor(10)
def f(x):
    return x**2
print(f(1), f(11))

#same
```
