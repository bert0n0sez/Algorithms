## Алгоритм Дейкстры

сделали табличку с 'уровнями'

Вычеркнули, и так далее для всех остальных

как можно ускорить? вместо массива с расстояниями до вершин мы сделаем КУЧУ, и вместо линейного времени мы получим логарифмическое врмея (на вытаскивание элемента)

асимптотика O(nlog(n))
Ищет кратчайшее расстояние от одной вершины для всех остальных, работает только для положительных весов, и с ориентированным и с неориентированным

```python
from random import randint

def read_graph():
    G = {}
    n, m = map(int, input().split())
    for _ in range(m):
        u, v = input().split()
        if u not in G:
            G[u] = {}
        if v not in G:
            G[v] = {}
        w = randint(1, 100)      #ВЕС, РАНДОМ ГОЙДА
        G[u][v] = w
        G[v][u] = w            #non orient dalbaeb

    return G



from heapq import heappush, heappop


def deykstra(G: dict, start: str):
    h = []
    heappush(h, (0, start))
    dist = {v: float('+inf') for v in G}      #сначала всем + бесконечность
    dict[start] = 0           #кроме начальной
    used = set()

    
    while len(h):
        w, v = heappop(h)             
        for heigh in G[v]:                #достали последнего - тк куча это мин, смотрим на его соседей, даем им новые (если меньше) веса
            if w + G[v][heigh] < dist[heigh]:
                dist[heigh] = w + G[v][heigh]
                heappush(h, (dist[heigh], heigh))
        used.add(v)
    return dist
```
