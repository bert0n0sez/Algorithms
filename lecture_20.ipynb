## Графы. Обход в глубину. Алгоритмы.

Алгоритм Косарайю - выделение сильных компонент связности. (Cильная связность - можем из любой точки попасть в любую). Граф ориентированный.


Компонента сильной связности - запускаем обход по инверсивному графу - справа - вершины куда можно попасть из икса, слева - откуда. Берем пересечение этих множеств - это компонента сильной связности.

Алгоритм:

1) берем первую вершину
2) начинаем обход в глубину
3) вычеркиваем то, куда попали
4) запускаем обход в глубину, из не вычеркнутой
5) получили последовательность вершин (в порядке как вычеркивали)
6) начинаем обход в глубину с конца - ходим в противоположную сторону
7) далее начинаем обход с первой невычеркнутой

те принцип - два раза делаем одно и то же: обход в глубину на исходном графе, и на инвертированном. Результат - выдает компоненты связности.

## Обход в ширину

1) стартуем с некоторой вершины
2) берем и обходим сразу всех соседей, приписали всем номера
3) берем след номер и обходим всех его соседей
4) ... и тд рекурсивно
5) ну и не ходим туда где цифра уже прописана


тут нужна очередь, нумеруем и по очереди берем
очередь в питоне? есть модуль queue - ориентирован на многопоточные штуки, есть свои минусы. Список - медленно, поэтому используеум collections.deque - двусторонняя очередь тип

```python
from collections import deque

def bfs(G, start):      #в ширину гойда
    d = deque()
    used = {start: 1}      #нумеруем уже посещенные
    d.append(start)      
    number = 2     
    while d:
        v = d.pop_lef()
        for n in G[v]:
            if n not in used:
                d.append(n)
                used[n] = number
                number += 1
    print(used)
```
